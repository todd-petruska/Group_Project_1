```python
# Checking the number of records
record_count18 = len(final18_df["Item_Number"].value_counts())
record_count19 = len(final19_df["Item_Number"].value_counts())
record_count20 = len(final20_df["Item_Number"].value_counts())
record_count21 = len(final21_df["Item_Number"].value_counts())
record_count22 = len(final22_df["Item_Number"].value_counts())

print(record_count18)
print(record_count19)
print(record_count20)
print(record_count21)
print(record_count22)
```

```python
# Flag duplicate records
dupcheck_18_df=final18_df.loc[final18_df.duplicated(subset=["Item_Number"])]
dupcheck_19_df=final19_df.loc[final19_df.duplicated(subset=["Item_Number"])]
dupcheck_20_df=final20_df.loc[final20_df.duplicated(subset=["Item_Number"])]
dupcheck_21_df=final21_df.loc[final21_df.duplicated(subset=["Item_Number"])]
dupcheck_22_df=final22_df.loc[final22_df.duplicated(subset=["Item_Number"])]
```

```python
# Keep only first row of duplicate Item Numbers
dupcheck_18_df.drop_duplicates(subset = ["Item_Number"], keep = 'first', inplace = True)
dupcheck_19_df.drop_duplicates(subset = ["Item_Number"], keep = 'first', inplace = True)
dupcheck_20_df.drop_duplicates(subset = ["Item_Number"], keep = 'first', inplace = True)
dupcheck_21_df.drop_duplicates(subset = ["Item_Number"], keep = 'first', inplace = True)
dupcheck_22_df.drop_duplicates(subset = ["Item_Number"], keep = 'first', inplace = True)
```

```python
# Rechecking the number of records after removing duplicated Item Numbers
nodup_record_count_18 = len(dupcheck_18_df["Item_Number"].value_counts())
nodup_record_count_19 = len(dupcheck_19_df["Item_Number"].value_counts())
nodup_record_count_20 = len(dupcheck_20_df["Item_Number"].value_counts())
nodup_record_count_21 = len(dupcheck_21_df["Item_Number"].value_counts())
nodup_record_count_22 = len(dupcheck_22_df["Item_Number"].value_counts())

print(nodup_record_count_18)
print(nodup_record_count_19)
print(nodup_record_count_20)
print(nodup_record_count_21)
print(nodup_record_count_22)
```

```python
# Set all five datasets together
five_yr_df=dupcheck_18_df.append([dupcheck_19_df, dupcheck_20_df, dupcheck_21_df, dupcheck_22_df],ignore_index=True)
five_yr_df.head()
```
